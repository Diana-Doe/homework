```python
import pandas as pd 
import json
with open('data.json', 'r', encoding='utf-8') as ff:
    data = json.load(ff)
```

```python
da = pd.concat({key: pd.DataFrame(val) for key, val in data['data'].items()}).reset_index()
del da['level_1']
da = da.rename(columns={"level_0": "date"}).set_index(['name'])
da
```

```python
categories, names, dates, prices, customers, rates = [], [], [], [], [], []
for _, row in da.iterrows():
    date = row.date
    name = row.name
    price = row.price
    customer = row.customers
    rate = row.rate
    for category in row.category:
        dates.append(date)
        names.append(name)
        prices.append(price)
        categories.append(category)
        customers.append(customer)
        rates.append(rate)
data = pd.DataFrame({
    "date" : dates,
    "category": categories,
    "name": names,
    "price" : prices,
    "customers" : customers,
    "rate" : rates
})
data = data.set_index(['date', 'name'])
data
```
